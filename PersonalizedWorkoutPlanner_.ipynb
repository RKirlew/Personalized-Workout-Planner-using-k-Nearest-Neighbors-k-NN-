```python
import pandas as pd
import random

fitness_levels = ['Beginner', 'Intermediate', 'Advanced']
goals = ['Weight Loss', 'Muscle Gain', 'Endurance', 'Flexibility']
workout_types = ['Cardio', 'Strength', 'Flexibility']

workout_recommendations = {
    'Cardio': ['Treadmill', 'Cycling', 'Jogging', 'Rowing', 'Swimming', 'HIIT', 'Jump Rope', 'Stair Climbing','LIIT'],
    'Strength': ['Squats', 'Deadlifts', 'Push-ups', 'Bench Press', 'Pull-ups', 'Dumbbell Rows', 'Overhead Press', 'Bicep Curls', 'Plank Holds','Hammer Curls','Lateral Raises'],
    'Flexibility': ['Yoga', 'Pilates', 'Stretching', 'Tai Chi', 'Foam Rolling', 'Barre Workouts', 'Dynamic Stretches', 'Mobility Drills']
}

data = []
for _ in range(200):
    age = random.randint(18, 60)
    fitness_level = random.choice(fitness_levels)
    goal = random.choice(goals)
    workout_type = random.choice(workout_types)
    recommended_workouts = random.sample(workout_recommendations[workout_type], 5)
    data.append([age, fitness_level, goal, workout_type, ", ".join(recommended_workouts)])

df = pd.DataFrame(data, columns=['Age', 'Fitness Level', 'Goal', 'Workout Type', 'Recommended Workouts'])

df.to_csv('expanded_workout_data.csv', index=False)
print("Expanded dataset created and saved as 'expanded_workout_data.csv'")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('expanded_workout_data.csv')
df_encoded = pd.get_dummies(df, columns=['Fitness Level', 'Goal', 'Workout Type'])

scaler = MinMaxScaler()
df_encoded['Age'] = scaler.fit_transform(df_encoded[['Age']])

X = df_encoded.drop('Recommended Workouts', axis=1)
y = df_encoded['Recommended Workouts']

vectorizer = CountVectorizer()
y_encoded = vectorizer.fit_transform(y).toarray()

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)
```

```python
new_data = {
    'Age': [24],
    'Fitness Level_Beginner': [0],
    'Fitness Level_Intermediate': [1],
    'Fitness Level_Advanced': [0],
    'Goal_Weight Loss': [0],
    'Goal_Muscle Gain': [1],
    'Goal_Endurance': [0],
    'Goal_Flexibility': [0],
    'Workout Type_Cardio': [0],
    'Workout Type_Strength': [1],
    'Workout Type_Flexibility': [0]
}

new_df = pd.DataFrame(new_data)

new_df = new_df[X_train.columns]

predicted_workouts = knn.predict(new_df)

predicted_workouts_decoded = vectorizer.inverse_transform(predicted_workouts)

print("Recommended Workouts:", predicted_workouts_decoded[0])
```
